### Can we make sense of simulation results in terms of basic physics of what's going on?

Simulations are a great tool to give us accurate results (albeit often at great computational expense), but they are not always great at giving us insights into key physics of the relevant processes.  Can we understand their results in terms of key underlying physical processes? 

A number of physical analytic and "semi-analytic" models exist for decribing the process of gravitational collapse of density perturbations. We will consider just a couple that are very useful and are most frequently encountered in the literature. 

### Spherical collapse model for the evolution of perturbations in expanding universe.  

 Although this is an example of a "spherical cow" model, 
this particular model turns out to be remarkably useful. 
  
<br>

### The spherical collapse model assumptions: monotonic density profile assumption

We will assume that density profile of the $\delta>0$ perturbation is monotonically increasing (or is constant) with decreasing $r$ (opposite assumption for $\delta<0$):

$$\rho(<r_1) \geq \rho(<r_2),\ \ \ \ \ \mathrm{for\ any}\ r_1<r_2\leq R_i$$

With this assumption, the enclosed mass $M(<r)$ stays constant during evolution of the perturbation. (Why?)

The fluctuation in which $\rho(<r)=\rm const$ for all $r\leq R_i$ is called the **top-hat fluctuation**.

The mass within a given radius will stay constant because shells that have larger density will always lag behind shells of lower density while shells expand, and will collapse faster than outer shells when shells contract (when they will collapse on the free-fall time scale that scales as $t_{\rm ff}\propto \rho^{-1/2}$).

Thus, the mass within $r$ for a density perturbation with such density profile will be conserved during collapse until radius of the perturbation becomes smaller than a certain radius where it meets collapsed shells of smaller radius that already went through $r=0$ point. 

### Then... 

If we consider a peak of *physical* radius $r$ enclosing mass $M=M(<r)$, the density within the *physical* radius $r$ within the perturbation and corresponding overdensity $\delta(<r)$ will be 

$$\rho=\frac{3M}{4\pi r^3};\ \ \ \ \  \delta=\frac{\rho}{\bar{\rho}_{\rm m}}-1.$$ 


### The mean density of the universe evolves as 

$$\bar{\rho}_{\rm m}(t)=\bar{\rho}_{\rm m0} a(t)^{-3},$$ 

where $\bar{\rho}_{\rm m0}$ is the mean matter density in the universe at the current time (redshift $z=0$) and $a(t)=(1+z)^{-1}$ is the expansion factor. 

### Thus, dimensionless density contrast evolves as

$$
1+\delta(t)=\frac{\rho}{\bar{\rho}_{\rm m}}=\frac{3M}{4\pi\bar{\rho}_{\rm m0}}\,\frac{a^3}{r^3}=R_{\rm L}^3\frac{a^3}{r^3},
$$

where $R_{\rm L}=[3M/(4\pi\bar{\rho}_{\rm m0})]^{1/3}$ is the *comoving Lagrangian radius* of the perturbation. 

### The physical radius of the perturbation thus evolves as 

$$r(t)=R_{\rm L}\, a(t)\,[1+\delta(t)]^{-1/3}.$$


Note that $R_{\rm L}$ is just a time-independent quantity defining the scale and mass of the perturbation. 

***Evolution of the rescaled quantity $r(t)/R_{\rm L}$ depends only on the background cosmology via $a(t)$ (the same for peaks of all masses) and overdensity $\delta$***: 

$$\frac{r(t)}{R_{\rm L}}=a(t)\,[1+\delta(t)]^{-1/3}.$$

$$r(t)=R_{\rm L} a(t)\,[1+\delta(t)]^{-1/3}.$$

<br>
The first implication of the above equation is rather trivial: 

* When $\delta=0$ (no perturbation),  $r=R_{\rm L}a(t)$ - i.e., region simply expands with the universe at the average rate. 

The second is less trivial...

* When fluctuation is negative, $-1<\delta_i<0$, $\ r_i>R_{\rm L} a(t_i)$ - i.e., at $t<t_i$ fluctuation was expanding faster than region of average density of the same mass. *It was getting larger with time in comoving units.*

* When fluctuation is positive, $\delta_i>0$, $r_i<R_{\rm L} a(t_i)$ - i.e., at $t<t_i$ fluctuation was expanding slower than the universe. *It was getting smaller with time in comoving units.*

### Consequence 1: 

Fraction of volume occupied by underdense regions will grow, while fraction of volume of overdense regions will decrease. 

Consequently, gravity and *gravitational instability will turn a Gaussian $\delta$ field into a non-Gaussian field.*

$$r(t)=R_{\rm L} a(t)\,[1+\delta(t)]^{-1/3}.$$

This equation can provide useful insights, but by itself it is not a solution for $\delta(t)$ because it combines two unknown functions, $\delta(t)$ and $r(t)$. 

### A number of approaches of solving this equation exists:

* linearizing time derivatives of this equation and solving the resulting linearized equations; this gives solution of the form (see A. Klypin's lectures and <a href="bg01_density_perturbations_evolution_linear.ipynb">bg01_density_perturbations_evolution_linear</a> notebook):
$\delta(t) \propto D_+(t, \Omega_{\rm m0}, \Omega_{\Lambda 0}),$<br>
where $D_+(t, \Omega_{\rm m0}, \Omega_{\Lambda 0})$ is the *linear growth factor* function that depends on the cosmological parameters describing background cosmology. $D_+$ can either be obtained analytically (for $\Omega_\Lambda=0$ cosmology) or numerically. 

    The fastest growth of $\delta$ occurs for $\Omega_{\rm m0}=1$ model: $\delta(t) \propto a(t)\propto t^{2/3}$

* solving the equation for $r(t)$ assuming spherical symmetry and $M(<r)=\rm const$ (*the spherical collapse model*)

### The spherical collapse model: assumptions

Neglecting pressure of matter and associated forces, as well as any force associated with vacuum energy, a particle at radius $r$ is accelerated only by the mass inside of $r$, i.e.:

$$\ddot{r}=-\frac{GM(<r)}{r^2},$$  

where $M(<r)=\rm const$ during evolution until the collapse to $r\approx 0$. This equation should be very familiar and has a simple analytic solution. 

For models with dark energy that has corresponding mean mass density $\rho_\Lambda$, the corresponding equation is 

$$\ddot{r}=-\frac{GM(<r)}{r^2} + \frac{8\pi G}{3}\,\rho_\Lambda r.$$  

In this case, the equation needs to be solved numerically. 

### Spherical collapse of $\delta>0$ perturbations: boundedness assumption

The first integral of our equation of motion, is total energy per unit mass (the sum of potential and kinetic energy per unit mass): 

$$E=\frac{\dot{r}^2}{2}-\frac{GM(<r)}{r}.$$

Given that here we are interested in the mass assembly of collapsed objects (galaxies and galaxy clusters), we will consider evolution of gravitationally bound, i.e. $E<0$ - i.e., perturbations that will eventually collapse.

### Key result of the spherical collapse model: linear overdensity at collapse time

Spherical collapse model calculations predict time $t_{\rm coll}$ at which initial overdensity $\delta_i(<r_i)$ will collapse, given that it is gravitationally bound. 

Given $t_{\rm coll}$ we can compute overdensity that would be predicted by linear regime evolution from $t_i$ to $t_{\rm coll}$:

$$\delta_c=\delta_i\, \frac{D_+(t_{\rm coll})}{D_+(t_i)}$$


### It turns out that $\delta_c$ only weakly depends on cosmological parameters

<br>
<center>

### Spherical collapse model of a spherical shell $\delta(<r)$


Assuming that $\delta_{i0}(<r_i)=\delta_i/D_+(t_i)$ is the initial overdensity profile linearly axtrapolated to $z=0$, $r_i$ is *comoving* radius of the initial shell, and growth function is normalized as $D_+(z=0)=1$, we can write:

$$\delta_{i0}(<r_i)D_+(t_{\rm coll})=\delta_{\rm c}$$

This is an implicit equation for the time of collapse $t_{\rm coll}$, which can be solved numerically for arbitrary cosmology for shells with different $r_i$. 

The value value of $\delta_c$ for $\Omega_{\rm m}=1$ cosmology can be understood as follows. In such case, the form of $r(t)$ is symmetric with respect to the turnaround time $t_{\rm ta}$ (maximum of $r(t)$): i.e., $t_{\rm coll}=2t_{\rm ta}$.

Thus, linear evolution from $t_{\rm ta}$ to $t_{\rm coll}$ would result in the increase of overdensity by a factor of 
$D_+(t_{\rm coll})/D_+(t_{\rm ta})=(t_{\rm coll}/t_{\rm ta})^{2/3}=2^{2/3}\approx 1.59$, where I used the fact that $D_+\propto a(t)\propto t^{2/3}$ in the $\Omega_{\rm m}=1$ cosmology. 

One could expect that at $t_{\rm ta}$ the linear regime breaks down and so linearly extrapolated overdensity at $t_{\rm ta}$ should be of order unity $\delta_{\rm ta}\sim 1$ and thus $\delta_c\approx 2^{2/3}\approx 1.6$. 

In fact, the exact calculation of the evolution equations in the case of 
$\Omega_{\rm m}(z)=1$ at the redshift of turnaround gives $\delta_{ta}=1.062$
(<a href="http://adsabs.harvard.edu/abs/1972ApJ...176....1G">Gunn & Gott 1972</a>), which gives
$\delta_c=\delta_{\rm ta} 2^{2/3}\approx 1.686$.


### For $\Omega_{\rm m}=1$, $\Omega_\Lambda=0$ cosmology solution is analytic: 

$D_+(z)=a(t)=(t/t_{\rm U})^{2/3}=(1+z)^{-1}$, so equation can be solved for $t_{\rm coll}$ or $z_{\rm coll}$ analytically, e.g.: 

$$z_{\rm coll} = \frac{\delta_i(<r_i)}{\delta_c} - 1$$

Note that even in the concordance $\Lambda$CDM model $\Omega_{\rm m}\approx 1$ at $z>2$ and the above equation provides a useful approximation for the redshift when a given radial shell will collapse.

Once $t_{\rm coll}$ or $z_{\rm coll}$ is computed for a given shell $r_i$, the mass it encloses

$$M(<r_i)=\frac{4\pi}{3}\,\bar{\rho}_{\rm m0}\,[1+\delta(<r_i)]\,r_i^3 $$

has collapsed. Vectors of $t_{\rm coll}(r_i)$ (or $z_{\rm coll}(r_i)$)  and $M(<r_i)$ for all shells comprise the mass assembly history of the peak. 

Time derivative of this function gives us *mass accretion rate* that such object experiences at any given time. 

$$\dot{M}_{\rm coll}(t)=\frac{dM_{\rm coll}(t)}{dt}=\frac{dM_{\rm coll}(z)}{dz}\,\frac{dz}{dt}.$$

According to the spherical collapse model ***mass accretion history of collapsing object simply reflects its initial $\delta(<r)$ profile.***

### Density profile taking into account existence of smaller $\delta$ sub-peaks

This can be done using a clever idea presented by <a href="http://arxiv.org/abs/1010.2539">Dalal et al. (2010)</a>.  
Details of the calculation, relevant equation, and example code computing this profile, can be found in the <a href="bg01_statistics_density_perturbations.ipynb">bg01_statistics_density_perturbations.ipynb</a> notebook. 

The idea is to compute an average value of the *maximum $\delta$* that $N=(R/r)^3$ subpeaks of radius $r$ can have within sphere of radius $R$, if the $\delta$ values of the subpeaks are randomly drawn from $p_{\rm G}(\delta_r \vert \delta_{R\ast})$.

 The main point to remember is that *unsmoothed density profile of a peak in a Gaussian field can be computed from the linear Gaussian field itself.* 

### Supremum statistics profile

If $x$ is a Gaussian random variable with pdf $p_{\rm G}(x)$ of zero mean and unit variance and we have $N$ samples $\{x_i\}$, the probability that a given $y$ is larger than any one $x_i$ is 
$$
P_1(y)=\int_{-\infty}^y\frac{dp_{\rm G}}{dx}dx = 1-\frac{1}{2}\mathrm{erfc}\left(\frac{y}{\sqrt{2}}\right),
$$

Assuming that $x_i$ samples are uncorrelated (not quite correct, but ignore correlations), the probability for $y$ to exceed all $N$ samples is: $P_{\rm N}(y) = [P_1(y)]^N.$ 

The differential pdf that $y$ is the supremum of $N$ samples of $x$ is $dP_N/dy$ and we can get an average value of $y$ in the usual way: 

$$\langle y\rangle = \int y\, \frac{dP_{\rm N}}{dy}\, dy = N\,\sqrt{\frac{2}{\pi}} \int y\,\left[1-\frac{1}{2}\mathrm{erfc}\left(y\right)\right]^{N-1} e^{-y^2}\, dy$$. 




### Supremum statistics profile

When we consider radius $r$ of the peak profile, we can think of $N\approx (R/r)^3$ sub-peaks of radius $r$ fitting within $R$. 

One of the sub-peaks has the largest $\delta$ and is located somewhere within sphere of radius $R$. We assume that it is this sub-peak that is identified as main progenitor in simulations. 

The sub-peaks have $\delta_r$ drawn from the Gaussian pdf $p_{\rm G}(\delta_r \vert \delta_{R\ast})$ for which we already know mean and variance. Thus, average *supremum* $\langle\delta_{r,\rm sup}\rangle$ can be computed using  $\langle y\rangle$ above as 

$$\langle\delta_{r,\rm sup}\rangle = \bar{\delta}(<r\vert\delta_{R\ast}) + \langle y\rangle\, \sigma_{\delta_r\vert \delta_{R\ast}} $$

### Ellipsoidal collapse model and corresponding collapse threshold

<a href="http://adsabs.harvard.edu/abs/1996ApJS..103....1B">Bond & Myers (1996)</a> carried out a systematic study of collapse of homogeneous ellipsoids of a given overdensity $\delta$ and trixial shape characterized by the three axes, $\lambda_1$, $\lambda_2$, $\lambda_3$ or shape parameters *ellipticity*, $e=(\lambda_1-\lambda_3)/(2\delta)$, and *prolaticity*, $p=(\lambda_1-2\lambda_2+\lambda_3)/(2\delta)$. 

The collapse model in this case instead of simply describing evolution of $r(t)$ as $\ddot{r}=-GM/r^2$, considers evolution of three axis of the ellipsoid, $r_1$, $r_2$, $r_3$, separately ($i=1,2,3$): 

$$\ddot{r}_i = -\frac{GM}{r_i^2} - S_i - \tau_i,$$

where $S_i$ is the term that describes effect of deviations of mass distribution from spherical symmetry, while $\tau_i$ describes effect of external tides on the evolution of the ellipsoid. The latter tends to slow down collapse of ellipsoid compared to spherical perturbation of the same $\delta$. 

<a href="http://adsabs.harvard.edu/abs/2001MNRAS.323....1S">Sheth, Mo & Tormen (2001,</a> hereafter SMT01) used results of such ellipsoidal collapse model and pdf of $e$ and $p$ for a Gaussian field based on Doroshkevich (1970) to show that linear overdensity at collapse time for homogeneous ellipsoid, $\delta_{\rm ec}$, is given by 

$$\delta_{\rm ec}(e,p)=\delta_{\rm sc}\,\left[1+\beta\left(5(e^2\pm p^2)\frac{\delta^2_{\rm ec}(e,p)}{\delta^2_{\rm sc}}\right)^\gamma\right],$$
where $\delta_{\rm sc}\approx 1.69$ is linear overdensity at collapse time predicted by spherical model. 

This expression can be used for individual peaks with known $e$ and $p$, but to develop a statistical threshold for collapse of ellipsoidal peaks, SMT01 used $p=0$, which corresponds to the peak of the prolaticity pdf, and the value of $e_{\rm mp}=(\sigma/\delta)\sqrt{5}$ that follows a maximum ridge of pdf $g(e,p=0\vert\delta)$. In this case the linear overdensity at collapse becomes:

$$\delta_{\rm ec}=\delta_{\rm sc}\left[1+\beta\left(\frac{\sigma^2}{\delta_{\rm sc}}\right)^\gamma\right]=\delta_{\rm sc}\left[1+\beta\nu^{-\gamma}\right].$$

The parameters $\beta=0.412$ and $\gamma=0.618$ chosen to provide best results (<a href="http://adsabs.harvard.edu/abs/2008MNRAS.388..638D">Desjacques 2008</a>).

Here is how the ellipsoidal collapse threshold compares to initial overdensities of peaks that collapse onto halos in numerical cosmological simulations of structure formation (Fig 3 of <a href="http://adsabs.harvard.edu/abs/2009ApJ...696..636R">Robertson et al. 2009</a>). The figure shows that the linearly extrapolated initial overdensity of peaks that collapse into halos by $z=0$ is generally larger than 1.69 predicted by the spherical collapse model, but is well described on average by the above expression for $\delta_{\rm ec}$.

### The takeaway points: 

* The mass accretion history $M(t)$ of collapsing dark matter halos *simply reflects the shape of the density profiles of their initial profiles, $\delta(<r)$, in the Gaussian $\delta$ field.*


* Each shell of the peak collapses at the time that is determined by its *linearly theory extrapolated* $\delta(<r)$ and collapse threshold that can be computed using simple spherical or ellipsoidal collapse model. 


* the shallow slope of $\delta(<r)$ at small $r/R_{\rm L}\ll 1$ results in rapid mass assembly at early times (high $z$), while steep slope of $\delta(<r)$ at $r\sim R_{\rm L}$ and larger results in slow mass assembly. This results in *two regimes of mass assembly*: 1) *fast mass accretion regime* and 2) *slow mass accretion regime.*


* As we will see later, imprint of these two regimes of total mass assembly is also apparent in the evolutionary histories of other galaxy properties, such as gas mass, stellar mass, star formation rate, etc.


### Mass accretion rate

Using simulation results and 

Starting with prediction for the total mass accretion rate, $\dot{M}$, as a function of object mass and time $t$ either taken directly from simulation or parametrized by an analytical model (e.g., such as those discussed in <a href="l01_peak_profile_MAH.ipynb">lecture 1</a> and the first practical exercise in <a href="ex01.ipynb">ex01</a> notebook).  

### Accretion rate of baryons

Let's assume that this total mass accretes onto galaxy as a mix of dark matter and diffuse baryon plasma (gas), with a mass fraction of baryons, $f_{\rm b,acc}$, so that baryon mass accretion rate is: 

$$\dot{M}_{\rm b,acc}= f_{\rm b,acc}\,\dot{M}=f_{\rm b,uni}\,f_{\rm b,in}\dot{M},
$$

where $f_{\rm b,uni}=\Omega_{\rm b}/\Omega_{\rm m}$ is "universal" baryon fraction (i.e., average over universe), which reflects the fact that overall baryons constitute a fraction of mass smaller or equal to this fraction. 

The factor $f_{\rm b,in}$ parametrizes possible enhancement or suppression of baryon accretion due to some physical process. 

### A simple "barebones" galaxy formation model

Assume that 1) baryons are accreted with universal baryon fraction (i.e., $f_{\rm b,in}=1$) in the form of gas and 2) after accretion this gas is delivered instantly to the interstellar medium 

$$
\begin{cases}
\dot{M}_{\rm g, ISM}= f_{\rm b,uni}\,f_{\rm b,in}\dot{M} - (1-\mathcal{R})\,\dot{M}_\star,\\
\dot{M}_\star =\frac{M_{\rm g, iSM}}{\tau_{\rm sf}} , \\
\end{cases}$$

where $\tau_{\rm sf}$ is *gas depletion time scale* on which ISM gas is converted into stars; in this simple model we will simply assume that $\tau_{\rm sf}=\rm const$. 

$\mathcal{R}$ is the fraction of newly formed stars that is returned back into the ISM via winds and supernovae explosions (we are assuming that this occurs instantly, using so-called *"instantaneous recycling approxination"*). The value of $\mathcal{R}$ depends on the initial mass function (IMF) of stars; we will assume a constant value characteristic of typical observed IMF: $\mathcal{R}=0.46$.

These equations can be numerically integrated as a system of ODEs. Such integration is implemented in <a href="codes/galaxy_model.py"><tt>codes/galaxy_model.py</tt></a> in python class <tt>simplest_model</tt>.

### Example of a process that can suppress gas accretion: UV/X-ray heating after reionization 

At $z_{\rm rec}\approx 1100$ the electrons and ions recombined, which allowed photons we know detect as the Cosmic Microwave Background (CMB) to propagate freely. 

Observations show, however, that by $z\approx 6$ intergalactic medium in the universe is ionized. So between $z=1100$ and $z\approx 6$ the gas must have been *re-ionized*. The process is called *reionization* and is almost certainly due to the UV photons emitted by young massive stars in the first galaxies at these high redshifts. 

### Re-ionization of the universe and heating of the intergalactic gas

The UV photons ionize neutral atoms and freed electrons carry extra energy which gets converted into gas temperature. Ionization by photons with spectrum typical of that of massive stars heats the gas to $T\approx 10000-20000 K$. This heating breaks self-similarity because it *introduces a mass scale* corresponding roughly to the *Jeans mass* of the integalactic gas and leads to suppression of gas accretion onto halos with mass smaller than the Jeans mass.


### The Jeans mass

This results in a characteristic halo mass scale below which gas accretion is suppressed that evolves with redshift. 

We can numerically write Jeans length for ideal ionized gas as the time scale over which the sound-crossing time and free-fall time are equal $\lambda_J/c_s\approx t_{\rm ff}$ or $\lambda_J\approx c_s\,t_{\rm ff}$

$$\lambda_J=c_s\sqrt{\frac{3\pi}{32G\rho}}=\sqrt{\gamma\,\frac{n_{\rm g}kT}{n_{\rm g}\mu m_{\rm p}}}\,\sqrt{\frac{3\pi}{32G\rho}}$$

where the last part assumes ideal ionized gas. Jeans mass can be defined as: 

$$M_J = \frac{4\pi}{3}\rho\,\left(\frac{\lambda_J}{2}\right)^3.$$



### The Jeans mass

For ionized gas $\mu\approx 0.6$, $\gamma=5/3$. Assuming that $\rho$ is the mean density of matter at redshift $z$: $\rho=(1+\delta)\,\Omega_{\rm m0}\rho_{\rm crit0}(1+z)^3$

$$M_J\approx 2.76\times 10^{8}\,M_\odot\,\left(1+\delta\right)^{-1/2}\,\left(\frac{1+z}{7}\right)^{-3/2}\,\left(\frac{\Omega_{\rm m0}}{0.3}\right)^{-1/2}\left(\frac{h}{0.7}\right)^{-1/2}\,\left(\frac{T}{10^4\ \rm K}\right)^{3/2} $$

The Jeans mass grows with time and for the same parameters at $z=0$ will be $7^{3/2}$ larger: $M_J\approx 5.11\times 10^9\, M_\odot$.


Inherit my <tt>simplest_model</tt> modell class into a new class (e.g., <tt>model_jeans_mass</tt>) and modify routine <tt>fg_in</tt>, which modulates accretion rate of baryon gas relative to the total mass accretion rate of matter, with a code that cuts off accretion of gas in objects that at $z\leq 6$ have mass smaller than the Jeans mass estimate above for $\delta=0$. 

```python
class model_jeans_mass(simplest_model):
    def __init__(self, *args, **kwargs):

        super(model_jeans_mass, self).__init__(*args, **kwargs)
        return
       
    def Mjeans(self, z):
        return 2.76e8/((1.+z)/7)**1.5 # Jeans mass at mean density for T=1e4 K and delta=0
    
    def fg_in(self,t):
        """
        implementing suppression of baryon accretion below Jeans mass
        """
        zd = self.cosmo.age(t, inverse=True)
        dummy = np.ones_like(zd)
        if zd<=6 and self.Mh <= self.Mjeans(zd):
            dummy = 0.0

        return dummy
```

### Suppression of gas accretion on small-mass halos and "filtering mass" 

The mathematical formalism describing effect of such re-heating on the evolution of perturbations is presented in <a href="http://adsabs.harvard.edu/abs/2000ApJ...542..535G">Gnedin (2000).</a> The quantitative effect on accretion of baryons on a collapsing peak is calibrated using
 numerical simulations of galaxy formation  (e.g., <a href="http://adsabs.harvard.edu/abs/2006MNRAS.371..401H">Hoeft et al 2006</a>, <a href="http://adsabs.harvard.edu/abs/2008MNRAS.390..920O">Okamoto et al 2008</a>).  f_{\rm b}(M,z)=\bar{f}_{\rm b}\left\{1+(2^{\alpha/3-1}-1)\left[\frac{M}{M_{\rm c}(z)}\right]^{-\alpha}\right\}^{-3/\alpha},
$$
 
with $\alpha=2$.

```python
from codes.variances import delta_variance, dvars, ddvars, ex_mean
from colossus.halo.mass_so import M_to_R
from colossus.lss import peaks

def compute_ave_dpro_for_deltaR(cosmo, MR, dRstar, ri, expro=False, verbose=False):
    rtol = 1.e-8

    RTH = peaks.lagrangianR(MR*cosmo.h) / (1. + dRstar)**(1./3.)
    if verbose: 
        zd = 0.
        print("R200m = %.3f kpc at z = %.3f"%(M_to_R(MR*cosmo.h, zd, '200m')/cosmo.h, zd))
        print("R_TH = %.3f comoving Mpc"%(RTH/cosmo.h))

    dRR = delta_variance(RTH, RTH, func=dvars, rtol=rtol, cosmo=cosmo)

    dlproave = np.zeros_like(ri)
    vardlproave = np.zeros_like(ri)

    for ir, rd in enumerate(ri):
        # <delta_r^2>
        var   = delta_variance(rd, rd, func=dvars, rtol=rtol, cosmo=cosmo)
        # <delta_r x delta_R>
        vardd = delta_variance(rd, RTH, func=ddvars, rtol=rtol, cosmo=cosmo)
        # average delta(<r) profile using equation above 
        dlproave[ir] = vardd / dRR * dRstar
        # variance about the average profile, can't be <0
        vardlproave[ir] = np.maximum(0., var - vardd**2 / dRR)
        
    if expro:
        for ir, rd in enumerate(ri):
            Nr = (RTH/rd)**3
            xex = 0.
            if Nr > 1:
                xex = ex_mean(Nr, rtol=rtol)
                dlproave[ir] = xex*np.sqrt(vardlproave[ir]) + dlproave[ir]

    return dlproave, vardlproave
```

```python
from scipy.interpolate import UnivariateSpline

def dpro_to_mah(cosmo, zmin, zmax, nzg, ri, dproi, delta_c=None):
    zg = np.linspace(zmax, zmin, nzg)
    # corresponding age of the universe at zg
    tg = cosmo.age(zg)
    #overdensity required to collapse by zg
    Dgr = cosmo.growthFactor(zg)
    # interpolate computed profile with cubic spline, so we can easily compute its derivative
    lDzsp  = UnivariateSpline(np.log10(Dgr), np.log10(1.+zg), s=0.)

    Mri = 4 * np.pi / 3. * (1. + dproi) * cosmo.rho_m(z=0.) * ri**3 * 1.e9 / cosmo.h
    zcr = np.maximum(10.**lDzsp(np.log10(delta_c/dproi)) - 1., -0.998)
    return zcr, Mri
```

```python
from codes.variances import delta_variance, dvars, ddvars, dddvars, ddddvars
from scipy.special import erfc
from codes.variances import ex_mean

from colossus.halo.mass_so import M_to_R
from colossus.lss import peaks

def compute_ave_dpro_for_deltaR_slope(cosmo, MR, dRstar, dddRstar, ri, expro=False, verbose=False):
    rtol = 1.e-9

    RTH = peaks.lagrangianR(MR*cosmo.h) / (1. + dRstar)**(1./3.)
    if verbose: 
        zd = 0.
        print("R200m = %.3f kpc at z = %.3f"%(M_to_R(MR*cosmo.h, zd, '200m')/cosmo.h, zd))
        print("R_TH = %.3f comoving Mpc"%(RTH/cosmo.h))

    Y = np.zeros(2); 
    Y[0] = dRstar; Y[1] = dddRstar # -6.5 for Via Lactea 2 specifically

    YY = np.zeros((2,2))
    YY[0,0] = delta_variance(RTH, RTH, func=dvars, rtol=rtol, cosmo=cosmo)
    YY[0,1] = delta_variance(RTH, RTH, func=dddvars, rtol=rtol, cosmo=cosmo)
    YY[1,0] = YY[0,1]
    YY[1,1] = delta_variance(RTH, RTH, func=ddddvars, rtol=rtol, cosmo=cosmo) 
    YYi = np.linalg.inv(YY)

    XY = np.zeros(2); 
    dlproave = np.zeros_like(ri)
    vardlproave = np.zeros_like(ri)

    for ir, rd in enumerate(ri):
        var = delta_variance(rd, rd, func=dvars, rtol=rtol, cosmo=cosmo)
        vardd = delta_variance(rd, RTH, func=ddvars, rtol=rtol, cosmo=cosmo)
        varddd = delta_variance(rd, RTH, func=dddvars, rtol=rtol, cosmo=cosmo)
    
        XY[0] = vardd; XY[1] = varddd
        dlproave[ir] = np.dot(XY,np.dot(YYi,Y))
        vardlproave[ir] = np.maximum(0., var - np.dot(XY,np.dot(YYi,XY.T)))

    if expro:
        for ir, rd in enumerate(ri):
            Nr = (RTH/rd)**3
            xex = 0.
            if Nr > 1:
                xex = ex_mean(Nr, rtol=rtol)
                dlproave[ir] = xex*np.sqrt(vardlproave[ir]) + dlproave[ir]

    return dlproave, vardlproave
```

```python
dRstar = 1.69; dddRstar = -3.
MR = 1.e12 
RL = peaks.lagrangianR(MR*cosmo.h) 
sigM = np.sqrt(delta_variance(RL, RL, func=dvars, rtol=1.e-8, cosmo=cosmo))
dRstar = delta_ce(cosmo,sigM,bec,gec)

dexproaves, vardexproaves = compute_ave_dpro_for_deltaR_slope(cosmo, MR, dRstar, dddRstar, ri, expro=True)
```
